```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
grid_u = np.linspace(-2.8, 2.8, 3)
grid_v = np.linspace(-1, 1, 2)
u, v = np.meshgrid(grid_u, grid_v)
x = np.ravel(u)
y = np.ravel(v)
plt.scatter(x, y)
```

```python
### create data X
grid_u = np.linspace(-2.8, 2.8, 30)
grid_v = np.linspace(-1, 1, 20)
u, v = np.meshgrid(grid_u, grid_v)
x = np.cos(np.ravel(u))
y = np.ravel(v)
z = np.sin(np.ravel(u))
X = np.array([x, y, z]).T
print(X.shape)

ax = plt.axes(projection='3d')
ax.scatter(x, y, z)
```

```python
from sklearn.decomposition import PCA
change = PCA(2) ### n_components=2
X_new = change.fit_transform(X)
plt.scatter(X_new[:,0], X_new[:,1])
```

## Manifold learning
PCA, MDS, Isomap, LLE, SpectralEmbedding

```python
from sklearn.manifold import MDS ### multidimensional scaling
change = MDS(2) ### n_components=2
X_new = change.fit_transform(X)
plt.scatter(X_new[:,0], X_new[:,1])
```

```python
from sklearn.manifold import Isomap 
change = Isomap(n_neighbors=5, n_components=2) ### n_neighbors=5, n_components=2
X_new = change.fit_transform(X)
plt.scatter(X_new[:,0], X_new[:,1])
```

```python
from sklearn.manifold import LocallyLinearEmbedding 
change = LocallyLinearEmbedding(n_neighbors=5, n_components=2) ### n_neighbors=5, n_components=2
X_new = change.fit_transform(X)
plt.scatter(X_new[:,0], X_new[:,1])
```

```python
from sklearn.manifold import SpectralEmbedding 
change = SpectralEmbedding(n_neighbors=5, n_components=2) ### n_neighbors=5, n_components=2
X_new = change.fit_transform(X)
plt.scatter(X_new[:,0], X_new[:,1])
```
